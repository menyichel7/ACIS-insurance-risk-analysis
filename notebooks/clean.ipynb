```python
import pandas as pd

# Load the actual/raw data
df = pd.read_csv('../data/MachineLearningRating_v3.csv', delimiter='|')

# Basic inspection
df.head()
df.info()
df.describe()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)

# Drop rows with missing values in key columns (adjusted column names)
df = df.dropna(subset=[
    "Gender", 
    "MaritalStatus", 
    "Product",            # Using 'Product' for PolicyType
    "VehicleType",        # Using 'VehicleType' for VehicleCategory
    "TotalPremium",       # Premium
    "TotalClaims"         # HasClaimed
])

# Fill missing categorical data where appropriate
df["Gender"] = df["Gender"].fillna("Unknown")
df["MaritalStatus"] = df["MaritalStatus"].fillna("Unknown")

# Create binary HasClaimed variable from TotalClaims
df["HasClaimed"] = df["TotalClaims"].apply(lambda x: 1 if x > 0 else 0)

# If needed, create Age from RegistrationYear (assuming 'current year' = 2015 based on data range)
import datetime
current_year = 2015
df["Age"] = current_year - df["RegistrationYear"]

# Quick check of cleaned data
print(df[["Age", "Gender", "MaritalStatus", "Product", "VehicleType", "TotalPremium", "HasClaimed"]].head())
```

```python
# Create binary claim flag
df["has_claim"] = df["HasClaimed"].apply(lambda x: 1 if x == "Yes" else 0)


```

```python
df.to_csv("cleaned_claims.csv", index=False)
```
